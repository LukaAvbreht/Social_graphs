```python
import numpy as np
import pandas as pd
import warnings
import ast
```

Create director file. We go through the movie data set and find all directors. We then find their average rating by taking the average rating of their movies and find their total vote count by summing the vote count of their movies.

```python
data_path = "the-movies-dataset/"

director_popularity_file = "director_pupularity.csv"
```

```python
# read datafiles
metadata = pd.read_csv(data_path+'movies_metadata.csv', encoding ='utf-8')
released = metadata.loc[metadata['status']=='Released']
cast_crew_data = pd.read_csv(data_path+'credits.csv', encoding ='utf-8')
```

```python
# get director of movie
def get_director(movie_crew):
    for crew_object in movie_crew:
        if (crew_object['job'] == 'Director'):
            return crew_object['name']
```

```python
# Initialize director dict

director_ratings = dict()
director_to_movies_id = dict()
```

```python
# go through the movies dataset
for ind,row in released.iterrows():
    try:
        movie_id = row["id"]
        vote_avrg = row["vote_average"]
        vote_count = row["vote_count"]
        movie_id = int(movie_id)

        crew = cast_crew_data.loc[cast_crew_data['id']==movie_id]["crew"].item()

        crew_object = ast.literal_eval(crew)
        director = get_director(crew_object)

        if director is None or director == "None":
            print("   There is no director for movie: {0}".format(row["title"]))
        else:
            # if this director has already been added
            if director in director_ratings:
                director_ratings[director] += ((vote_avrg,vote_count),)
                director_to_movies_id[director] += (movie_id,) 
            else:
                # if this director has not been added yet
                director_ratings[director] = ((vote_avrg,vote_count),)
                director_to_movies_id[director] = (movie_id,) 
    except:
        print("Failed in row with index: {0}".format(row["id"]))
```

```python
data_to_save = list()
for director in director_ratings.keys():
    director_data = list(director_ratings[director])
    director_movies = list(director_to_movies_id[director])
    
    avrg_score = round(sum((i for i,j in director_data))/len(director_data),3)
    score_count = sum((j for i,j in director_data))

    data_to_save.append([score_count,avrg_score,director,director_movies])

data_to_save.sort(reverse=True)

df_data = pd.DataFrame(data_to_save, columns = ['vote_count', 'vote_average',"director","work"]) 
with open(data_path+director_popularity_file,"w") as ooutput:
    df_data.to_csv(ooutput,index=True)
```

Create the smaller datafiles based on directors voting

```python
# Read ratings file
director_popularity = pd.read_csv('../../director_popularity.csv', encoding ='utf-8')
len(director_popularity.index)
```

```python
# only directors with a vote count over 500
director_popularity_small = director_popularity.loc[director_popularity['vote_count'] >= 500]

# get all movies from these directors
l = list(director_popularity_small['work'])
movies = []
for sublist in l:
    sublist = ast.literal_eval(sublist)
    for item in sublist:
        movies.append(item)
```

```python
len(movies)
```

```python
# get relevant rows from movies metadata
metadata = pd.read_csv('../../the-movies-dataset/movies_metadata.csv', encoding ='utf-8')
str_list = [str(i) for i in movies] 
metadata_small = metadata.loc[metadata['id'].isin(str_list)]
```

```python
# get relevant rows from ratings
ratings = pd.read_csv('./../../the-movies-dataset/ratings.csv', encoding ='utf-8')
warnings.filterwarnings(action='once')
ratings_small = ratings.loc[ratings['movieId'].isin(movies)]
```

```python
# get relevant rows from credits
credits = pd.read_csv('./../../the-movies-dataset/credits.csv', encoding ='utf-8')
credits_small = credits.loc[credits['id'].isin(movies)]
```

```python
# save to files
metadata_small.to_csv('../../smaller_dataset/movies_metadata_short.csv')
ratings_small.to_csv('../../smaller_dataset/ratings_short.csv')
credits_small.to_csv('../../smaller_dataset/credits_short.csv')
```
